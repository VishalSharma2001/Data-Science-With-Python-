```python
import numpy as np
import matplotlib.pyplot as plt
x=np.arange(1,11)
y=[2,4,5,7,6,8,9,10,12,6]
#x=np.arange(1,11)
#y=np.arange(11,21)
plt.scatter(x,y,label="Star",marker="*",s=40,color="b")
plt.plot(x,y)
plt.grid()
plt.legend()
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("Scatter plot")
plt.show()
```
