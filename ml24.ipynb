```python
import matplotlib.pyplot as plt
import csv
x=[]
y=[]
with open("test.txt","r") as csvfile:
    plots=csv.reader(csvfile)
    for col in plots:
        x.append(col[0])
        y.append(col[1])
plt.plot(x,y)
plt.grid()
plt.show()
```
