```python
import matplotlib.pyplot as plt
x=[1,2,3,4,5]
y=[45,38,25,15,8]
lab=["one","two","third","four","five"]
plt.bar(x,y,tick_label=lab,width=.8,color=["red","orange","Black"])
plt.xlabel("x-axis")
plt.ylabel("t-axis")
plt.title("bar plot")
plt.show()
```
