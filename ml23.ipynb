```python
import matplotlib.pyplot as plt
import numpy as np
x1=np.arange(0,2*np.pi,.01)
y1=np.cos(x1)
y2=np.sin(x1)
plt.plot(x1,y1,label="Cos")
plt.plot(x1,y2,label="Sin")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("Sine & Cosine Functions")
plt.legend()
plt.show()
```
